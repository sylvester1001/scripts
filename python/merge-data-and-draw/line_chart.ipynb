```python
import matplotlib.pyplot as plt
import pandas as pd
import os

# outputCSV
output_csv_files = [file for file in os.listdir('./output') if file.endswith('.csv')]

# CSV
for output_csv_file in output_csv_files:
    # 
    output_csv_path = os.path.join('./output', output_csv_file)

    # CSV
    data = pd.read_csv(output_csv_path)

    # 
    x = data.iloc[:, 0]  # 
    y = data.iloc[:, 1:6]  # 

    # 
    plt.plot(x, y)
    plt.legend(data.columns[1:6])  # 
    title=os.path.splitext(output_csv_file)[0]
    plt.title(title)

    output_fig_path = "./fig"
    os.makedirs(output_fig_path, exist_ok=True)
    # 
    output_csv_filename = os.path.splitext(output_csv_file)[0]  # 
    fig_save_name = os.path.join(output_fig_path, f"{output_csv_filename}.png")  # 
    plt.savefig(fig_save_name)

    # 
    plt.show()

```
